```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Load data
df = pd.read_csv("winedata_coord.csv")
df.head(2)
```

```python
#Create dataframe for wines in 2013

df_2013Wine = df.loc[df["Year"]==2013]

df_2013Wine.head()
```

```python
# The value_counts method counts unique values in a column
dfTopVariety = df["Variety"].value_counts()
dfTopVariety.head(5)
```

```python
df_PinotNoir = df_2013Wine.loc[df["Variety"]=="Pinot Noir"]
df_PinotNoir.head()
```

```python
# Create a group based on the values in the 'Country' column
country_group = df_PinotNoir.groupby('Country')

# Count how many times each country appears in our group
count_country = country_group['Country'].count()

count_country
```

```python
# Create a bar chart based off of the group series from before
count_chartcountry = count_country.plot(kind='bar')

# Set the xlabel and ylabel using class methods
count_chartcountry.set_xlabel("Countries")
count_chartcountry.set_ylabel("Number of Pinot Noir Wine")

# Save our graph and show the grap
plt.tight_layout()
plt.savefig("PinotNoir2013.png")
plt.show()
```

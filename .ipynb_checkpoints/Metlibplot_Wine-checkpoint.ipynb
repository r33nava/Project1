```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("winedata_coord.csv")

df.head(5)
```

```python
# Reference multiple columns within a DataFrame
df[["Province", "Country","Price","Grade"]].head(5)
```

```python
# Collect only those projects that were hosted in the US
US_Wineries = df.loc[df["Country"]=="US"]

US_Wineries.head()
```

```python
# Reference multiple columns within a US DataFrame
US_Wineries[["Province", "Country","Price","Grade"]].head(5)
```

```python
# Collect only those projects that were hosted in the US
NJ_Wine = US_Wineries.loc[US_Wineries["Province"]=="New Jersey"]

NJ_Wine.head(100)
```

```python
# Set x axis and tick locations
x_axis = np.arange(len(NJ_Wine))
tick_locations = x_axis

#tick_locations = [value for value in x_axis]
```

```python
# Create a list indicating where to write x labels and set figure size to adjust for space
plt.figure(figsize=(20,3))

plt.bar(x_axis, NJ_Wine["Grade"], color='r', alpha=0.75, align="center")

plt.xticks(tick_locations, NJ_Wine["Price"], rotation="vertical")

#For David... 
ticks = plt.bar(x_axis, NJ_Wine["Grade"], color='r', alpha=0.5, align="center")

def labelbar(ticks):
    for tick in ticks:
        height = tick.get_height()
        plt.text(tick.get_x() + tick.get_width()/2, height, int(height), ha='center', va='bottom')
        
labelbar(ticks)
```

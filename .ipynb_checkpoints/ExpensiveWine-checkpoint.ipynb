```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("winedata_coord.csv")

df.head(5)
```

```python
df_ExpensiveWine = df.loc[df["Price"]>=156]

df_ExpensiveWine.head()
```

```python
a = df_ExpensiveWine["Grade"]

b = df_ExpensiveWine["Price"]
```

```python
# Tells matplotlib that we want to make a scatter plot
# The size of each point on our plot is determined by their x value
plt.scatter(a, b, marker="o", facecolors="green", edgecolors="black", alpha=0.75)
```
